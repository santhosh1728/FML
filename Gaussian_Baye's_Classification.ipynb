```python
df.drop(labels='Id',axis=1,inplace=True)
```

```python
columns = df.columns.tolist()
```

```python
Classes = (df[
  'Species'].unique()
)
```

```python
Classes
```

```python
result = df.loc[df['Species'] == Classes[0]], columns[0]
```

```python
def get_pdf_par(attribute,Classes):
  result = (df.loc[df['Species'] == Classes])[attribute]
  mean = np.mean(result)
  var = np.std(result)**2
  return mean,var
```

```python
def get_likelihood(value,mean,var):
  r = 1/(np.sqrt(2*np.pi*var))*np.exp(-(value-mean)**2/(2*var))
  return r
```

```python
df.head()
```

```python
T = {'SepalLengthCm':5,'SepalWidthCm':2,
     'PetalLengthCm':3.5,'PetalWidthCm':1}
```

```python
T.keys()
```

```python
S = []
for j in Classes:
  L = 0
  for i in T.keys():
    m,v = get_pdf_par(i,j)
    likelihood = np.log(get_likelihood(T[i],m,v))
    L = likelihood
  S.append(L)
```

```python
ind_1 = np.where(S == np.max(S))
predicted_class = Classes[ind_1[0][0]]
```
